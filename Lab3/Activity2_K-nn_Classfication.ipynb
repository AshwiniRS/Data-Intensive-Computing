<h3> KNN Function </h3>

Defining function for sampling the data and applying knn function and finding out the misclassification rate.

```r
knnDataFn = function(n.points,sampling.rate){

    num.test.set.labels <- n.points * (1 - sampling.rate)
    training <- sample(1:n.points, sampling.rate * n.points,replace=FALSE)
    train <- subset(data[training, ], select = c(Age,Credit.amount))
    testing <- setdiff(1:n.points, training)
    test <- subset(data[testing, ], select = c(Age,Credit.amount))
    cl <- data$Credit.Risks[training] #this is the subset of labels for the training set
    true.labels <- data$Credit.Risks[testing]
    
    resultdf = data.frame(k=double(),misclassification.rate=double())
    for(k in 1:20) {
        predicted.labels <- knn(train,test,cl,k)
        num.incorrect.labels<-sum(predicted.labels !=true.labels)
        misclassification.rate<-num.incorrect.labels/num.test.set.labels
        newRow = data.frame(k,misclassification.rate)
        resultdf = rbind(resultdf,newRow)
        
    }
    return(list(resultdf,train,cl))
}
```

This will be or test data to check the output for different k values.

```r
test.data=data.frame(Age=60,Credit.amount=1199)
```

Creating data frame to store minimum classification rate and corresponding k.

```r
min.observation = data.frame(k=double(),misclassificaton.rate=double(),sampling.rate=double())
```

<h2><b>For 70%-30% partition of data as training and test</b></h2>

```r
sampling.rate = 0.7
resultdf_0.7 = knnDataFn(522,0.7)

#retrieving data from returned variable 
cl=data.frame(resultdf_0.7[3])
cl=data.frame(cl=cl[,])
cl=cl[,1]
train=data.frame(resultdf_0.7[2])
train=data.frame(Age=train$Age,Credit.amount=train$Credit.amount) 
kmisclass = data.frame(resultdf_0.7[1])
# kmisclass
```

```r
#finding min
mindf = kmisclass[which.min(kmisclass$misclassification.rate),]
newRow = data.frame(mindf$k,mindf$misclassification.rate,sampling.rate)
min.observation = rbind(min.observation,newRow)
mindf
```

```r
kmin=mindf$k
kmin
```

```r
sprintf("For k which has least misclassification rate:%i",kmin)
x=knn(train,test.data,cl,k=kmin)
y=x[1]
sprintf("Predicted credit risk level: %s",y)
sprintf("For k=4")
x=knn(train,test.data,cl,k=4)
y=x[1]
sprintf("Predicted credit risk level: %s",y)
```

```r
for(kitr in 1:20){
    print(kitr)
    x=knn(train,test.data,cl,k=kitr)
    y=x[1]
    print(y)
}
```

<h2>Using Normalization</h2>

Let us normalize data and get all fields in numeric. We will bring the data in 1 to 10 range. For that we will assign number to a every category if it is categorical column. 

• Age - we can take mod of the entries.<br>
• Sex - only two values - we can assign 1=female and 2=male <br>
• Job - 0,1,2,3 <br>
• Housing - own=1, free=2, rent=3 <br>
• Savings.account - NA=1, little=2, quite rich=3, rich=4, moderate=5 <br>
• Checkings.account - little=1, moderate=2, NA=3, rich=4 <br>
• Duration - we can take mod of the entries <br>
• Purpose - radio/TV=1, education=2, Furniture/equipment=3, Car=4, Business=5, Domestic Appliances=6, Repairs=7, Vacation/Others=8 <br>
• Credit Risks = 1,2

<B>Defining Functions</B>

```r
modFn = function(x){
    mod = x%%10
    return (mod)
}
```

```r
modFn_100 = function(x){
    mod = x%%100
    return (mod)
}
```

Applying Functions

```r
data$Age <- lapply(data$Age, function(x) sapply(x,modFn))
data$Duration <- lapply(data$Duration, function(x) sapply(x,modFn))
data$Credit.amount <- lapply(data$Credit.amount, function(x) sapply(x,modFn_100))
data$Credit.amount <- lapply(data$Credit.amount, function(x) sapply(x,modFn))
```

Factorizing the data.

```r
data$Sex = factor(data$Sex, labels = c("male"="1", "female"="2"))
data$Housing = factor(data$Housing, labels = c("own"="1", "free"="2", "rent"="3"))
data$Saving.accounts = factor(data$Saving.account, labels = c("little"="1", "moderate"="2", "quite rich"="3", "rich"="4"))
data$Checking.account = factor(data$Checking.account, labels = c("little"="1", "moderate"="2", "rich"="3"))
data$Purpose = factor(data$Purpose, labels = c("radio/TV"="1","euctaion"="2","furniture/equipment"="3","car"="4","buisnes"="5",
                                              "domestic appliances"="6","repairs"="7","vactions/others"="8"))
```
