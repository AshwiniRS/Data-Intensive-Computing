Reading the csv files.

```r
playersData <- read.csv(file="player.csv", header=TRUE)
teamsData <- read.csv(file="team.csv", header=TRUE)
```

<h3>Q1: Which are the top 10 players by overall rating? </h3>

```r
rating_sort=(playersData[order(-playersData$rating),]) #sort dataframe according to rating in descending order
top10rating_ans=rating_sort[1:10,c('player_api_id','player_name','rating')] #select top 10 rows
top10rating_ans
```

<h3>Q2: Which are the top 10 goalkeepers by sum of gk attributes? </h3>
