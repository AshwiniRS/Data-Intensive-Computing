For middle younger age group, the count gradually increases accroding to income and then decreases irrespective of sex. Middle older and older people have average counts with maximum of them giving agreement over the opinion. There is not much difference between the count of male and female numbers. Overall we can conclude that maximum people think online dating is good to meet new people irrespective of age,sex,income.
